```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, SimpleRNN, Dense, Dropout

# Load your dataset
df = pd.read_csv("E:/vscode/CryptoCast_bitcoin_project/Bitcoin Historical Data (1).csv")
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df = df.dropna(subset=['Date','Price']).sort_values(by='Date')

prices = df['Price'].to_numpy().reshape(-1,1)
scaler = MinMaxScaler()
prices_scaled = scaler.fit_transform(prices)

seq_length = 60
X, y = [], []
for i in range(len(prices_scaled) - seq_length - 7):
    X.append(prices_scaled[i:i+seq_length])
    y.append([prices_scaled[i+seq_length], prices_scaled[i+seq_length+2], prices_scaled[i+seq_length+6]])
X, y = np.array(X), np.array(y)

# CNN model
cnn_model = Sequential([
    Conv1D(16, 3, activation='relu', input_shape=(seq_length,1)),
    MaxPooling1D(pool_size=2),
    Dropout(0.2),
    Dense(1)
])
cnn_model.compile(loss="mse", optimizer="adam")
cnn_model.fit(X, y[:,0], epochs=20, batch_size=32)
cnn_model.save("cnn_model.h5")

# LSTM model
lstm_model = Sequential([
    LSTM(32, input_shape=(seq_length,1)),
    Dropout(0.2),
    Dense(1)
])
lstm_model.compile(loss="mse", optimizer="adam")
lstm_model.fit(X, y[:,1], epochs=20, batch_size=32)
lstm_model.save("lstm_model.h5")

# RNN model
rnn_model = Sequential([
    SimpleRNN(16, input_shape=(seq_length,1), activation='tanh'),
    Dropout(0.2),
    Dense(1)
])
rnn_model.compile(loss="mse", optimizer="adam")
rnn_model.fit(X, y[:,2], epochs=20, batch_size=32)
rnn_model.save("rnn_model.h5")
```
